```python
import io
import gzip
import pandas as pd

from garmin_fit_sdk import Decoder, Stream
```

```python
file_name = 'tp-3801599.2026-03-01-16-35-50-965Z.GarminPing.AAAAAGmkauYYHXMq.FIT.gz'
```

```python
gz = gzip.open(file_name, 'rb').read()
```

```python
with open('extracted_file.fit', 'wb') as f_out:
    f_out.write(gz)
```

```python
stream = Stream.from_file('extracted_file.fit')
decoder = Decoder(stream)
messages, errors = decoder.read()

# print(errors)
# print(messages)
```
